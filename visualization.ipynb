```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("fashion-mnist_train.csv")
```

```python
trainX = train.loc[:,train.columns != 'label']
```

```python
trainY = train.loc[:,train.columns == 'label']
```

```python
random_samples = trainX.sample(n = 100).to_numpy()
```

```python
rows = 10
cols = 10
axes=[]
fig=plt.figure()
for a in range(rows*cols):
    axes.append( fig.add_subplot(rows, cols, a+1) )
    plt.imshow(random_samples[a].reshape(28,28),cmap=plt.get_cmap('gray'))
plt.show()
```

```python
rows = 2
cols = 5
axes=[]
fig=plt.figure()
for a in range(rows*cols):
    axes.append( fig.add_subplot(rows, cols, a+1) )
    plt.hist(random_samples[a], bins=32, range=(0, 255))
plt.show()
```
